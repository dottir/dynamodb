```python
# query in DynamoDB example

import boto3
from boto3.dynamodb.conditions import Key

def query():
    dynamodb = boto3.resource('dynamodb')
    
    table = dynamodb.Table('loan-2014')      
    
    resp = table.query(
        KeyConditionExpression=Key('uuid').eq('11514680')
    )
                
    return resp         
```
